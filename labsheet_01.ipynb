```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[4,5,6]
plt.plot(x,y)
plt.title('info')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.figure(figsize=(500,500))
```

```python
x=[20,30,40]
plt.pie(x)
plt.show()
```

```python
x=[10,20,30]
mylabels=["ferrari","porsche","bugati"]
plt.pie(x,labels=mylabels)
plt.show()
```

```python
x=[10,20,30]
mylabels=["ferrari","porsche","bugati"]
mycolours=["black","green","red"]
plt.pie(x,labels=mylabels)
plt.pie(x,colors=mycolours)
plt.show()
```

```python
import matplotlib.pyplot as plt
y=[10,20,30]
mylabels=["ferrari","porsche","bugati"]
myexplode=[0.2,0.2,0.2]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```

```python
import numpy as np
x=np.array(["a","b","c","d"])
y=np.array([3,8,1,10])
plt.bar(x,y)
plt.show()
```

```python

x=np.array(["a","b","c","d"])
y=np.array([3,8,1,10])
mycolours=["red","black","blue","green"]
plt.barh(x,y,color=mycolours)
plt.show()
```

```python
x=[2,4,6,8]
y=[1,4,5,7]
mycolours=["blue","yellow","red","green"]
plt.scatter(x,y,color=mycolours,marker="*")
plt.show()
```

```python
x=[2,4,6,8]
y=[1,4,5,7]
size=[150,300,460,590]
mycolours=["blue","yellow","red","green"]
plt.scatter(x,y,color=mycolours,marker="*",s=size,alpha=0.1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.subplot(2,2,1)
x=np.array(["a","b","c","d"])
y=np.array([3,8,1,10])
plt.bar(x,y)
plt.subplot(2,2,2)
x=np.array(["a","b","c","d"])
y=np.array([3,8,1,10])
plt.scatter(x,y)
plt.subplot(2,2,3)
x=np.array(["a","b","c","d"])
plt.pie([3,8,1,10])
plt.subplot(2,2,4)
x=np.array(["a","b","c","d"])
y=np.array([3,8,1,10])
plt.plot(x,y)
plt.show()
```
