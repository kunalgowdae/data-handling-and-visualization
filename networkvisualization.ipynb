```python
import networkx as nx
import matplotlib.pyplot as plt
g=nx.Graph()
g.add_nodes_from([1,2,3,4,5])
g.add_edges_from([(1,2),(1,3),(2,3),(3,4),(5,4)])
pos=nx.spring_layout(g)
nx.draw(g,pos,with_labels=True,node_color='blue',node_size=1400,edge_color='black',linewidths=1,font_size=15)
plt.show()
```
